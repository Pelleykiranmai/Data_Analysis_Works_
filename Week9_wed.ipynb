```python
import numpy as np
import matplotlib.pyplot as plt

# ------------------- Task 1: Simulate Daily Emails -------------------
print("=== Task 1: Simulating Daily Email Counts ===")

# Parameters
lam = 10      # average number of emails per day
days = 30     # one month of workdays

# Generate Poisson-distributed data
emails_per_day = np.random.poisson(lam, days)

# Compute summary statistics
mean_emails = np.mean(emails_per_day)
var_emails = np.var(emails_per_day)
std_emails = np.std(emails_per_day)

# Display stats
print(f"λ (expected mean): {lam}")
print(f"Simulated Mean: {mean_emails:.2f}")
print(f"Variance: {var_emails:.2f}")
print(f"Standard Deviation: {std_emails:.2f}")
print(f"Simulated Daily Email Counts: {emails_per_day.tolist()}\n")

# Plot histogram
plt.figure(figsize=(10,5))
plt.hist(emails_per_day, bins=range(min(emails_per_day), max(emails_per_day)+2),
         color='PINK', edgecolor='black', alpha=0.7)
plt.axvline(lam, color='red', linestyle='--', label=f'Expected Mean = {lam}')
plt.title("Daily Email Counts (λ = 10) — 30-Day Simulation")
plt.xlabel("Number of Emails Received per Day")
plt.ylabel("Frequency (Days)")
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

# Interpretation
print("Interpretation:")
print("------------------------------------------------")
print(f"- The mean number of emails per day ≈ {mean_emails:.1f}, close to λ = 10.")
print("- The histogram peaks near 10 and shows a slight right skew, typical of Poisson data.")
print("- This suggests email arrival follows a random process with moderate variability.")
print("- Some days have more or fewer emails, but most days cluster around the mean.\n")


# ------------------- Task 2: Exploring Effect of λ -------------------
print("=== Task 2: Exploring Effect of Mean (λ) ===")

lambda_values = [2, 5, 10, 15, 25]
size = 1000  # simulate for 1000 days per scenario

plt.figure(figsize=(12,8))

for i, lam in enumerate(lambda_values):
    data = np.random.poisson(lam, size)
    plt.subplot(3, 2, i+1)
    plt.hist(data, bins=range(0, max(data)+2), color='SKYBLUE', edgecolor='black', alpha=0.7)
    plt.axvline(lam, color='red', linestyle='--', label=f"Mean = {lam}")
    plt.title(f"λ = {lam}")
    plt.xlabel("Emails per Day")
    plt.ylabel("Frequency")
    plt.legend()
    plt.grid(alpha=0.25)

plt.tight_layout()
plt.show()

# Interpretation of λ effect
print("Interpretation:")
print("------------------------------------------------")
print("- For λ = 2: Distribution is highly right-skewed (many days with very few emails).")
print("- For λ = 5: Still skewed but smoother.")
print("- For λ = 10 and 15: Distribution starts centering, variability becomes symmetric.")
print("- For λ = 25: Almost bell-shaped (approaching normal distribution).")
print()
print("Real-world meaning:")
print("- Small λ (e.g., 2 or 5) → employees receive few emails; communication is light.")
print("- Larger λ (e.g., 15 or 25) → busy roles or departments, frequent correspondence,")
print("  and higher communication intensity — requiring more time management and support.")
print()
print("Conclusion:")
print("As λ increases, the relative randomness decreases and daily workload becomes more predictable.")
```
