```python
#Arrays
# we create arrays using numpy module
import numpy as np
a1=[10,20,30,40]
a2=[[1,2,],[4,5,]]
print(a1)
print(a2)
a=np.array(a1)
b=np.array(a2)
c=np.array([[10,20],[30,40]])
print(a)
print(b)
print(c)
print(b+c)
print(c-b)
print(5*c)
print(b*c)
print(np.matmul(c,b))
d=np.array([[1,-2,3],[-4,5,6]])
print(np.transpose(d))
print(np.linalg.inv(c))
print(d[1][1])
```
