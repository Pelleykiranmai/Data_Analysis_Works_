```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parameters
n = 100
p_values = [0.2, 0.4, 0.8]  # Different system accuracies
simulations = 10000         # Number of simulations for visualization

plt.figure(figsize=(15, 5))

for i, p in enumerate(p_values, 1):
    # Simulate outcomes
    data = np.random.binomial(n, p, simulations)

    # Mean & Variance (theoretical)
    mean = n * p
    var = n * p * (1 - p)

    # Plot histogram
    plt.subplot(1, 3, i)
    plt.hist(data, bins=20, density=True, alpha=0.6, color="skyblue", edgecolor="black")

    # Overlay exact binomial distribution (PMF)
    x = np.arange(0, n+1)
    plt.plot(x, binom.pmf(x, n, p), 'r-', lw=2, label="Binomial PMF")

    # Mark mean & variance
    plt.axvline(mean, color="green", linestyle="--", label=f"Mean = {mean:.1f}")
    plt.title(f"Binomial Simulation (p={p})")
    plt.xlabel("Correct Diagnoses")
    plt.ylabel("Probability")
    plt.legend()

plt.tight_layout()
plt.show()
```
