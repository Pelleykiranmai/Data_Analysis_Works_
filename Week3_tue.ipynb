```python
import matplotlib.pyplot as plt

# Example data (replace with your real dataset)
offer_counts = [2, 1, 0, 4, 3, 6, 5, 7]
buy_counts   = [0, 2, 1, 3, 4, 2, 6, 3]
labels       = [0, 0, 0, 0, 1, 1, 1, 1]  # 0: Legitimate, 1: Spam

# Split the data
x_legit = [x for x, l in zip(offer_counts, labels) if l == 0]
y_legit = [y for y, l in zip(buy_counts, labels) if l == 0]
x_spam  = [x for x, l in zip(offer_counts, labels) if l == 1]
y_spam  = [y for y, l in zip(buy_counts, labels) if l == 1]

plt.scatter(x_legit, y_legit, c='green', marker='^', label='Legitimate')
plt.scatter(x_spam, y_spam, c='orange', marker='s', label='Spam')

plt.title("Distribution of Offer & Buy Word Usage in Emails")
plt.xlabel('Count of "offer"')
plt.ylabel('Count of "buy"')
plt.legend()
plt.show()
```

```python
import numpy as np

# Example data
hours_studied = np.array([2, 4, 6, 8, 10])
test_scores   = np.array([55, 60, 70, 85, 95])

# Define a threshold for "high" score
high_score_threshold = 80

# Mask for high and low scores
high = test_scores >= high_score_threshold
low  = test_scores < high_score_threshold

plt.scatter(hours_studied[low], test_scores[low], c='blue', marker='o', label='Score < 80')
plt.scatter(hours_studied[high], test_scores[high], c='red', marker='*', label='Score â‰¥ 80')

plt.title("Impact of Study Hours on Test Performance")
plt.xlabel('Hours Studied')
plt.ylabel('Test Score')
plt.legend()
plt.show()
```
