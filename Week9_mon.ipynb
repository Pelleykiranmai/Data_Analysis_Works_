```python
import numpy as np
import matplotlib.pyplot as plt

# ------------------- Task 1 -------------------
# Parameters
lam = 12      # mean number of visitors per minute
minutes = 60  # simulate for 1 hour

# Simulate minute-wise traffic using Poisson distribution
traffic = np.random.poisson(lam, minutes)

# Compute summary statistics
mean_traffic = np.mean(traffic)
var_traffic = np.var(traffic)
print("=== Task 1: Minute-wise Traffic Simulation ===")
print(f"Lambda (expected mean): {lam}")
print(f"Simulated Mean: {mean_traffic:.2f}")
print(f"Variance: {var_traffic:.2f}")
print(f"Std Deviation: {np.std(traffic):.2f}")
print()

# Plot histogram of visitor counts
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.hist(traffic, bins=range(min(traffic), max(traffic)+2),
         color='skyblue', edgecolor='black', alpha=0.7)
plt.title("Histogram: Visitors per Minute ( = 12)")
plt.xlabel("Number of Visitors")
plt.ylabel("Frequency")

# Line plot for minute-wise arrivals
plt.subplot(1,2,2)
plt.plot(range(1, minutes+1), traffic, marker='o', color='orange')
plt.axhline(y=lam, color='red', linestyle='--', label=f"Expected mean = {lam}")
plt.title("Minute-wise Traffic Pattern (1 Hour)")
plt.xlabel("Minute")
plt.ylabel("Visitors per Minute")
plt.legend()
plt.tight_layout()
plt.show()

# Interpretation
print("Interpretation:")
print("------------------------------------------------")
print(f"- The average number of visitors per minute  {mean_traffic:.1f}, close to  = 12.")
print("- The traffic fluctuates around the mean  some minutes have more, some fewer visitors.")
print("- This indicates natural variation in Poisson arrivals, but overall traffic is fairly stable.")
print("- In practice, this means the server load is relatively predictable with mild randomness.\n")


# ------------------- Task 2 -------------------
# Effect of Marketing Campaign
lam_before = 10
lam_during = 25
minutes = 60

traffic_before = np.random.poisson(lam_before, minutes)
traffic_during = np.random.poisson(lam_during, minutes)

# Compute statistics
mean_before = np.mean(traffic_before)
mean_during = np.mean(traffic_during)

print("=== Task 2: Effect of Marketing Campaign ===")
print(f"Before Campaign: ={lam_before}, Mean={mean_before:.2f}")
print(f"During Campaign: ={lam_during}, Mean={mean_during:.2f}")
print()

# Plot comparison
plt.figure(figsize=(10,5))
plt.hist(traffic_before, bins=range(0, max(traffic_during)+2),
         alpha=0.6, label=f"Before (={lam_before})", color='blue', edgecolor='black')
plt.hist(traffic_during, bins=range(0, max(traffic_during)+2),
         alpha=0.6, label=f"During (={lam_during})", color='orange', edgecolor='black')
plt.title("Visitor Distribution: Before vs During Campaign")
plt.xlabel("Number of Visitors per Minute")
plt.ylabel("Frequency")
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

# Line comparison
plt.figure(figsize=(10,5))
plt.plot(range(1, minutes+1), traffic_before, label='Before Campaign', marker='o', alpha=0.7)
plt.plot(range(1, minutes+1), traffic_during, label='During Campaign', marker='x', alpha=0.7)
plt.axhline(lam_before, color='blue', linestyle='--')
plt.axhline(lam_during, color='orange', linestyle='--')
plt.title("Minute-wise Traffic Comparison (Before vs During Campaign)")
plt.xlabel("Minute")
plt.ylabel("Visitors per Minute")
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

# Interpretation
print("Interpretation:")
print("------------------------------------------------")
print("- During the campaign,  increases from 10 to 25  meaning average traffic more than doubles.")
print("- The distribution becomes wider and more symmetric as  increases (variance  ).")
print("- Higher  implies:")
print("    Higher mean visitor rate  more server requests per minute.")
print("    Larger fluctuations  servers must handle higher peak loads.")
print("    Greater demand on bandwidth, caching, and response time management.")
print("- Therefore, marketing campaigns should be coordinated with IT teams to ensure adequate server capacity.")
```
